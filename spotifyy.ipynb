```python
#importing the datasets and loading them
import pandas as pd
import numpy as np
high=pd.read_csv("high_popularity_spotify_data.csv")
low=pd.read_csv("high_popularity_spotify_data.csv")
```

```python
#know the data summary,
#preview of actual raw data from "high" dataframe
high.info()
high.head()
```

```python
#know the data summary,
#preview of actual raw data from "low" dataframe
low.info()
low.head()
```

```python
#handling missing values
#using is NAN or NA and summing them
high.isna().sum()
```

```python
#handling missing values
#using is NAN or NA and summing them
low.isna().sum()
```

```python
#dropping unnecessary columns from low dataframe
col_low_drop=low.drop(["track_href","analysis_url","uri","id","track_album_id","playlist_id","track_id","type"],axis=1,inplace=True)
```

```python
#dropping unnecessary columns from high dataframe
col_high_drop=high.drop(["track_href","analysis_url","uri","id","track_album_id","playlist_id","track_id","type"],axis=1,inplace=True)
```

```python
high.drop_duplicates()
low.drop_duplicates()
```

```python
# convertion date into just year of released.
#song duration converting from milliseconds to minutes (high)
high['track_album_release_date'] = pd.to_datetime(high['track_album_release_date'], errors='coerce')
high['year_released'] = high['track_album_release_date'].dt.year
high['song_duration'] = high['duration_ms'] / 60000

```

```python
# convertion date into just year of released (low)
#song duration converting from milliseconds to minutes.
low['track_album_release_date'] = pd.to_datetime(low['track_album_release_date'], errors='coerce')
low['year_released'] = low['track_album_release_date'].dt.year
low['song_duration'] = low['duration_ms'] / 60000
```

```python
#converting track popularity into a seperate column popularity
high['popular_song'] = np.where(high['track_popularity'] >= 50, 1, 0)
low['popular_song']  = np.where(low['track_popularity'] >= 50, 1, 0)
```

```python
dropping_date_time_pop_high=high.drop(["track_album_release_date","duration_ms","track_popularity"],axis=1,inplace=True)
dropping_date_time_pop_low=low.drop(["track_album_release_date","duration_ms","track_popularity"],axis=1,inplace=True)
```

```python
#concating two dataframe for better analysis
high['popularity_level'] = 'high'
low['popularity_level'] = 'low'

combined = pd.concat([high, low], ignore_index=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=combined, x='danceability', hue='popularity_level', kde=True)
plt.show()
```
